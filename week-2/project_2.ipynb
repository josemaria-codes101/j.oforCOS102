```python
import math

print("Select the equation to solve")
print("1. Quadratic Equation")
print("2. Cubic Equation")
print("3. Quartic Equation")

choice = int(input("Enter your choice: "))

# QUADRATIC
if choice == 1:
    print("Quadratic Equation: Ax^2 + Bx + C = 0")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))

    D = B**2 - 4*A*C

    if D > 0:
        root1 = (-B + math.sqrt(D)) / (2*A)
        root2 = (-B - math.sqrt(D)) / (2*A)
        print("Roots are:", root1, "and", root2)

    elif D == 0:
        root = -B / (2*A)
        print("Root is:", root)

    else:
        print("Complex roots")


# CUBIC
elif choice == 2:
    print("Cubic Equation: Ax^3 + Bx^2 + Cx + D = 0")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))

    print("Approximate real roots:")

    x = -100
    while x <= 100:
        f = A*x**3 + B*x**2 + C*x + D
        if abs(f) < 0.1:
            print(round(x,2))
        x += 0.1


# QUARTIC
elif choice == 3:
    print("Quartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")

    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))

    print("Approximate real roots:")

    x = -100
    while x <= 100:
        f = A*x**4 + B*x**3 + C*x**2 + D*x + E
        if abs(f) < 0.1:
            print(round(x,2))
        x += 0.1

else:
    print("Invalid choice")
```
