```python
import tkinter as tk
from tkinter import ttk, messagebox
```

```python

# Parent Class
class Zenith:
    def mutual_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"
        ]

    def unique_services(self):
        return []


# ------------------ Child Classes ------------------
class RetailBanking(Zenith):
    def unique_services(self):  # Polymorphism (override)
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and savings"
        ]


class GlobalBanking(Zenith):
    def mutual_services(self):  # Global has no mutual from list, override
        return []

    def unique_services(self):  # Polymorphism
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]


class CommercialBanking(Zenith):
    def unique_services(self):  # Polymorphism
        return [
            "Advisory services"
        ]


# GUI
def display_services():
    name = name_entry.get()
    division = division_combo.get()

    if not name or not division:
        messagebox.showwarning("Input Error", "Please enter all fields")
        return

    # Object creation
    if division == "Retail Banking":
        employee = RetailBanking()
    elif division == "Global Banking":
        employee = GlobalBanking()
    elif division == "Commercial Banking":
        employee = CommercialBanking()
    else:
        return

    services = employee.mutual_services() + employee.unique_services()

    output_text.delete("1.0", tk.END)
    output_text.insert(tk.END, f"Employee: {name}\nDivision: {division}\n\nServices:\n")

    for service in services:
        output_text.insert(tk.END, f"- {service}\n")


```

```python
# ------------------ GUI Setup ------------------
root = tk.Tk()
root.title("Zenith Bank Services")
root.geometry("500x400")

# Name Input
tk.Label(root, text="Employee Name:").pack(pady=5)
name_entry = tk.Entry(root, width=40)
name_entry.pack(pady=5)

# Division Dropdown
tk.Label(root, text="Division:").pack(pady=5)
division_combo = ttk.Combobox(root, values=[
    "Retail Banking",
    "Global Banking",
    "Commercial Banking"
])
division_combo.pack(pady=5)

# Button
tk.Button(root, text="Show Services", command=display_services).pack(pady=10)

# Output Box
output_text = tk.Text(root, height=12, width=60)
output_text.pack(pady=10)

root.mainloop()
```
