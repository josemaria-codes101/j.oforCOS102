```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
# Clean the names
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
```

```python
# Clean the review numbers
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df.isna().sum()
```

```python
# Visualization 1: Shows spread of price
sns.histplot(df['price'])
```

Insight 1: 
The distribution of prices is highly right-skewed, with most listings concentrated at lower price ranges and a few listings with very high prices. This indicates the presence of few outlieers and majority of users prefer budget friendly accommodations.

```python
# Visualization 2: Shows how location affects price
sns.boxplot(x='neighbourhood_group', y='price', data=df)
```

Insight 2: 
Manhattan has the highest median prices. While neighborhoods like Staten Island and Bronx have significantly lower prices  

```python
# Visualization 3: Shows most common listing type
sns.countplot(x='room_type', data=df)
```

Insight 3: 
Entire home/apt and private rooms are commonly used rooms with the former being the most common type while shared rooms are significantly the least common
