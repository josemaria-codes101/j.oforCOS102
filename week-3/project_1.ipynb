```python
import tkinter as tk
from tkinter import messagebox as msgbox
import csv

# Load employees from CSV
def load_employees():
    employees = []
    with open("GIG-logistics(in).csv", newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
            employees.append(row)
    return employees

# Check employee
def check_employee():
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    found = False
    department_members = []

    for emp in employees:
        emp_name = emp["Name"].strip().lower()
        emp_dept = emp["Department"].strip().lower()

        if emp_name == name and emp_dept == dept:
            found = True

        if emp_dept == dept:
            department_members.append(emp["Name"])

    if found:
        msg = f"Welcome {name.title()}!\n\nOther members in {dept.title()}:\n"
        msg += "\n".join(department_members)
        msgbox.showinfo("Success", msg)
    else:
        msgbox.showerror("Error", "Employee does not exist.")

# Load data
employees = load_employees()

# GUI setup
root = tk.Tk()
root.title("GIG Logistics App")
root.geometry("400x250")

# Labels
tk.Label(root, text="Enter Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

# Button
tk.Button(root, text="Check Employee", command=check_employee).pack(pady=10)

# Run app
root.mainloop()
```
